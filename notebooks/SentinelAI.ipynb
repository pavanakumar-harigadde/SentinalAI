**8. Feature Engineering**

```python


#feature 1 : High Sensitivity
master_df["high_sensitive"] = (
    master_df["resource_sensitivity"] == "high" 
).astype(int)

#Feature 2 : Medium Sensitivity
master_df["medium_sensitive"] = (
    master_df["resource_sensitivity"] == "medium"
).astype(int)

#Feature 3: Night activity
master_df["night_access"] = (
    master_df["time_classification"] == "night"
).astype(int)

#Feature 4: Weekend Activity
master_df["weekend_access"] = (
    master_df["time_classification"] == "weekend"
).astype(int)

#Feature 5: Unusual Hours
master_df["unusual_hours"] = (
    master_df["time_classification"] == "unusual"
).astype(int)

#Feature 6: Failes access
master_df["failed_access"] =(
    master_df["status"] == "failure"
).astype(int)

#Feature 7: Dormant account (Stale users)
master_df["inactive_30"] = (
    master_df["days_inactive"] >= 30
).astype(int)

master_df["inactive_60"] =(
    master_df["days_inactive"] >= 60
).astype(int)


#visualize again
master_df.head()
```

**9. creating Risk score here.**

```python
master_df["risk_score"] =0

#Sensitivity
master_df["risk_score"] += master_df["high_sensitive"] * 25
master_df["risk_score"] += master_df["medium_sensitive"] * 20

#Time Context
master_df["risk_score"] += master_df["night_access"] * 20
master_df["risk_score"] += master_df["weekend_access"] * 15
master_df["risk_score"] += master_df["unusual_hours"] * 10

#Account activity 
master_df["risk_score"] += master_df["inactive_30"] * 10
master_df["risk_score"] += master_df["inactive_60"] * 20

#Access Status
master_df["risk_score"] += master_df["failed_access"] * 15
```

```python
# Time Bsed New Features are adding here

master_df["timestamp"] = pd.to_datetime(master_df["timestamp"])
master_df["hire_date"] = pd.to_datetime(master_df["hire_date"])

# Day of month
master_df["day_of_month"] = master_df["timestamp"].dt.day

# Month-end indicator
master_df["is_month_end"] = (
    master_df["day_of_month"] >= 28
).astype(int)

# Employee tenure
master_df["tenure_days"] = (
    master_df["timestamp"] -
    master_df["hire_date"]
).dt.days

# New hire flag
master_df["new_hire"] = (
    master_df["tenure_days"] <= 30
).astype(int)
```

```python
#Adjustments in score for special cases
def context_adjustment(row):

    adjustment = 0

    # Finance month-end work
    if (
        row["department"] == "Finance"
        and row["resource"] == "GL_System"
        and row["is_month_end"] == 1
    ):
        adjustment -= 10

    # Security team working on SIEM
    if (
        row["department"] == "Security"
        and row["resource"] == "SIEM"
    ):
        adjustment -= 10

    # Service accounts doing API jobs
    if (
        row["privilege_level"] == "service-account"
        and row["action"] == "api_call"
    ):
        adjustment -= 10

    # New employees
    if row["new_hire"] == 1:
        adjustment -= 5

    return adjustment
```

```python
#Applying Context Adjustments to columns
master_df["context_adjustment"] = master_df.apply(
    context_adjustment,
    axis=1
)

master_df["raw_risk_score"] = (
    master_df["raw_risk_score"]
    + master_df["context_adjustment"]
)
```

```python
#Re-normalizing risk score after adding adjustments to the risk score
master_df["risk_score"] = (
    master_df["raw_risk_score"]
    / master_df["raw_risk_score"].max()
) * 100

master_df["risk_score"] = master_df["risk_score"].round(0)
```

```python
#Recalculating Risk scores
def classify_risk(score):

    if score >= 80:
        return "Critical"

    elif score >= 60:
        return "High"

    elif score >= 40:
        return "Medium"

    else:
        return "Low"


master_df["risk_level"] = master_df["risk_score"].apply(
    classify_risk
)
```

**13. Isolation Forest**

```python
#Features for Isolation Forest
iso_df = master_df.copy()

iso_df["action_code"] = iso_df["action"].astype("category").cat.codes
iso_df["resource_code"] = iso_df["resource"].astype("category").cat.codes
iso_df["sensitivity_code"] = iso_df["resource_sensitivity"].astype("category").cat.codes
iso_df["privilege_code"] = iso_df["privilege_level"].astype("category").cat.codes
iso_df["inactive_days"] = iso_df["days_inactive"]
iso_df["status_code"] = iso_df["status"].astype("category").cat.codes
iso_df["time_code"] = iso_df["time_classification"].astype("category").cat.codes
iso_df["department_code"] = iso_df["department"].astype("category").cat.codes


features = iso_df[
[
    "action_code",
    "resource_code",
    "sensitivity_code",
    "status_code",
    "time_code",
    "department_code",
    "privilege_code",
    "days_inactive",
    "tenure_days"
]
]
```

```python
#TRaining with IF
from sklearn.ensemble import IsolationForest

iso = IsolationForest(
    contamination=0.1,
    random_state=42
)

iso.fit(features)

master_df["ml_anomaly"] = iso.predict(features)
```

```python
#Adding new feature unauthorized acccess
def unauthorized_access(row):

    allowed = str(row["systems_access"]).split("|")

    resource_map = {
        "Customer_Vault":"PROD_DB",
        "HRIS":"AD",
        "Admin_Console":"ADMIN_SYS",
        "SIEM":"SIEM",
        "GL_System":"PROD_DB",
        "Email_Archive":"EMAIL"
    }

    expected = resource_map.get(
        row["resource"],
        None
    )

    if expected and expected not in allowed:
        return 1

    return 0
```

```python
master_df["unauthorized_access"] = master_df.apply(
    unauthorized_access,
    axis=1
)
```

**14. Explainations**

```python
def generate_reason(row):

    reasons = []

    if row["resource_sensitivity"] == "high":
        reasons.append("Accessed high sensitivity resource")

    if row["time_classification"] in [
        "night",
        "weekend",
        "unusual_hours"
    ]:
        reasons.append("Off-hours activity detected")

    if row["days_inactive"] > 90:
        reasons.append("Dormant account activity")

    if row["new_hire"] == 1:
        reasons.append("New employee activity")

    if row["ml_anomaly"] == -1:
        reasons.append("Behavioral anomaly detected")

    if row["action"] == "admin_operation":
        reasons.append("Administrative action performed")

    if (row["action"] == "export_data" and row["resource_sensitivity"] == "high"):
        reasons.append("High-volume export from sensitive resource")

    elif row["action"] == "export_data":
        reasons.append("Data export activity detected")

    return " | ".join(reasons)
```

```python
master_df["investigation_reason"] = (
    master_df.apply(generate_reason, axis=1)
)
```

```python
def recommendation(row):

    if row["risk_level"] == "Critical":
        return "Immediate investigation and temporary account suspension"

    elif row["risk_level"] == "High":
        return "Review access logs and manager approval"

    elif row["risk_level"] == "Medium":
        return "Monitor activity"

    else:
        return "No action required"
```

```python
master_df["recommendation"] = (
    master_df.apply(recommendation, axis=1)
)
```

```python
def alert_title(row):

    if row["risk_level"] == "Critical":
        return "CRITICAL SECURITY ALERT"

    elif row["risk_level"] == "High":
        return "HIGH RISK ACCESS ALERT"

    elif row["risk_level"] == "Medium":
        return "MEDIUM RISK ACTIVITY"

    return "LOW RISK EVENT"
```

```python
master_df["alert_title"] = master_df.apply(
    alert_title,
    axis=1
)
```

```python
def generate_explanation(row):

    reasons = []

    reasons.append(
        f"User {row['username']} from {row['department']} performed {row['action']} on {row['resource']}."
    )

    if row["unauthorized_access"] == 1:
        reasons.append(
            "The resource is not present in the user's approved access profile."
        )

    if row["resource_sensitivity"] == "high":
        reasons.append(
            "The accessed resource contains highly sensitive information."
        )

    if row["night_access"] == 1:
        reasons.append(
            "Access occurred during night hours."
        )

    if row["weekend_access"] == 1:
        reasons.append(
            "Access occurred during a weekend."
        )

    if row["unusual_hours"] == 1:
        reasons.append(
            "Activity occurred outside normal working schedule."
        )

    if row["new_hire"] == 1:
        reasons.append(
            "The employee is a recent hire and has limited historical behavior."
        )

    if row["inactive_60"] == 1:
        reasons.append(
            "The account had been inactive for an extended period."
        )

    if row["is_month_end"] == 1 and row["department"] == "Finance":
        reasons.append(
            "Month-end finance processing was detected and risk was reduced accordingly."
        )

    if row["ml_anomaly"] == -1:
        reasons.append(
            "Behavior was flagged by the Isolation Forest anomaly detection model."
        )

    if row["actual_night"] == 1:
        reasons.append(
            "Access occurred during night hours."
    )

    if row["actual_weekend"] == 1:
        reasons.append(
            "Access occurred during weekend."
    )

    if row["new_hire"] == 1:
        reasons.append(
            f"Employee tenure is only {int(row['tenure_days'])} days."
    )

    if row["days_inactive"] > 30:
        reasons.append(
            f"Account inactive for {row['days_inactive']} days."
    )

    reasons.append(
        f"Final risk score: {row['final_risk_score']:.0f}/100."
    )

    return " ".join(reasons)
```

```python
master_df["explanation"] = master_df.apply(
    generate_explanation,
    axis=1
)
```

```python
def business_impact(row):

    if row["resource"] == "Customer_Vault":
        return "Potential exposure of customer data."

    elif row["resource"] == "HRIS":
        return "Potential exposure of employee records."

    elif row["resource"] == "GL_System":
        return "Potential impact on financial information."

    elif row["resource"] == "Admin_Console":
        return "Potential compromise of privileged systems."

    return "Potential security risk requires review."
```

```python
master_df["business_impact"] = master_df.apply(
    business_impact,
    axis=1
)
```
